## Algorithm: Linear Regression

need to use few linear regression libraries.

2
- linear regression library from sklearn
- import cross val score(going to do lot of train and test split)



```python
#CROSS Validation:

#Entire dataset - 100 records

#  ______________________________
#|                               |
 # ______________________________

#5 cross validation

#1st CV: 

#test data and remaining all training data

 #______________________________
#| Test |        Training       |
 # _____________________________

#2nd CV:

 #______________________________
#| Test | Test |  Training       |
 # _____________________________

#3rd CV:

#...

#so, will do cross validation 5 times taking diff combination of train and test data.
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

#intialize
lin_reg = LinearRegression()
#Applying cross validation
#CV- we divide train and test in such a way that every combination of train and 
#test is taken by the model and who's ever accuracy is better than entire thing is combined.

mse = cross_val_score(lin_reg, X, y, scoring='neg_mean_squared_error', cv=5)
#parameters-model-linear reg, X, y
#we are not doing train test split specifically here.
#we are giving entire X and y
#Based on that i'm going to do cross validation
#Can also do train test split intialling and then give x train and y train to do the cross
#validation.
```

Best practice:
Doing train test split and then only giving train data to do the cross validation.

scoring can be use MSE(mean squared error) and negMSE
can find all of this in sklearn page of cross val score.

Cv - cross validation value
here it is 5.

So, here as we are doiung 5 cross validation mse going to be 5 values.
|

```python
print(mse)
#all diff MSE-as we are doing Cv as 5
```

```python
mse_avg = np.mean(mse)
#average of all mean squared errors.
```

```python
print(mse_avg)
```

```python
#Neg as we used -ve mse.
#but considering only mse
#it is 0.558

#This is how we do cross validation.
```

## Ridge Regression:

```python
#hyperparameter tuning in ridge regression.

#2 diff libraries:
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
#This is for doing hyper parameter tuning.(by GridSearchCV)



#in order to get good value - we do hyperparameter tuning.

#model:
ridge = Ridge()

#opening sklearn ridge
#parameters
#1. alpha - learning rate (best parameter to perform hyperparameter tuning)

#2. max iterations - how many times to change theta1 value to get the right value.
```

```python
#Defining paramters:
#in dictionaries:
params = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}

#in grid search CV - they take combination of alpha value and wherever the model performs well
# and gives it as selected bestfit parameter.

ridge_regressor = GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv=7)
#parameters: model, parameters defined, scoring(neg mean squared error)
```

```python
ridge_regressor.fit(X, y)
#Can also do train test split and can do this process of model fit on train X and train y.
```

```python
print(ridge_regressor.best_params_)  #This gives best parameter(for alpha)
print(ridge_regressor.best_score_)  #This gives best score(error best)
```

bcz of ridge regression. Negative mean square error has definetely became better.

-0.5582901717686543 - for linear regression
    
-0.5548535736454416 - for ridge regression

So, should select linear regression as it performing better.

but, ridge reduce overfitting.(cannot use as performance is becoming very bad)

Now , we can try lasso regression.


## Lasso Regression:

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso = Lasso() #intialization of model.

params_l = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]} #defining parameters.

#cross validation:
lasso_reg = GridSearchCV(lasso, params_l, scoring='neg_mean_squared_error', cv=7)

#fitting the model:
lasso_reg.fit(X, y)
```

```python
print(lasso_reg.best_params_)
print(lasso_reg.best_score_)
```

-0.5548535736454416 - for ridge regression

-0.5550162785002432 - for lasso regression

## Logistic Regression:

```python
# can also do grid search cv.
```

```python
from sklearn.linear_model import LogisticRegression


#for logistic we solve using classification problem
#so, lets take new dataset:

#Dataset:
from sklearn.datasets import load_breast_cancer
```

```python
data_f = load_breast_cancer()
```

```python
data_f
```

```python
dataset_bc = pd.DataFrame(data_f['data'], columns=data_f['feature_names'])
```

```python
dataset_bc #independent features
```

```python
dataset_bc['Result']=data_f['target'] #Dependent feature
```

```python
dataset_bc
```

```python
#creating seperate data frames for independent and dependent features.

X = pd.DataFrame(data_f['data'], columns=data_f['feature_names'])
```

```python
X
```

```python
y = pd.DataFrame(data_f['target'])
```

```python
y.columns= ['Result']
```

```python
y
```

```python
#check if the output/dependent feature is balanced or imbalanced:

y['Result'].value_counts()

#it gives count of 1's and 0's
```

```python
#This is probably a balanced dataset.
```

```python
#Train Test split:

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=47)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
#Algorithm: LOgistic Regression

#parameters:
#1. penalty - to add L2 penalty or L1 penalty
#2. C - Inverse of regularisation strength.(1/lambda)
#3. class_weight - for not balanced dataset. Can apply weights to class.

#learn by 2 ways -probablistic and geometry.
```

```python
params = [{'C': [1, 5, 10]}, {'max_iter':[100, 150]}]

#here, playing with two parameters(C, max_iter)
#max_iter - specifically for grid search CV
```

```python
#Defining model:

model1=LogisticRegression(C=100,max_iter=100) # default values
#later on can apply gridsearch CV

model=GridSearchCV(model1, param_grid=params, scoring='f1', cv=5)
#since, it is classification problem. not pretty sure that whether 
#true+ve or true-ve is imporatnt

#Can use f1 scoring(It is basically parametric term-nothing but performance metrics)
```

```python
#So, THis is the model with respect to GridSearchCV
model
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
#96%

#still testing with test data

y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
#Can apply confusion matrix after getting prediction values:

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
#confusion matrix with respect to predicted and actual values

confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test, y_pred)

#this gives accuracy for the model.
```

```python
#for detailed precision, recall and all the scores:

print(classification_report(y_test, y_pred))
```

```python
#Since, this is balanced dataset-the performance will be best.
```
