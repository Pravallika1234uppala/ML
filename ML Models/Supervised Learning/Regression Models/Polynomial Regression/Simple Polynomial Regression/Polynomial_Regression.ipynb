## Polynomial Regression:


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Simple Polynomial Regression
```

```python
df = pd.read_csv("Ice_cream selling data.csv")
```

```python
#with Simple Linear Regression:
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0], df.iloc[:,1],test_size=0.4,random_state=42)
```

```python
X_train = X_train.to_frame()
```

```python
X_test = X_test.to_frame()
```

```python
#Simple Linear Regression:
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
```

```python
#Scaling the values:
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
#on training dsata -fit
#on testing data - transform
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
#training model:
regression_model.fit(X_train,y_train)
```

```python
y_pred=regression_model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
Rscore = r2_score(y_test, regression_model.predict(X_test))
print(Rscore)
```

```python
#less accuracy for linear reg
#visualizing model:
plt.subplot(1,2,1)
plt.scatter(X_train, y_train, c="g")
plt.plot(X_train, regression_model.predict(X_train), c="b")
plt.xlabel("Training Input")
plt.ylabel("Training Output")
plt.title("Training Best Fit line-LR")
plt.subplot(1,2,2)
plt.scatter(X_test, y_test, c="r")
plt.plot(X_test, regression_model.predict(X_test), c="y")
plt.xlabel("Testing Input")
plt.ylabel("Testing Output")
plt.title("Testing Best Fit line-LR")
```

```python
#Finding Error for Linear Regression:
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root mean Squared Error:", rmse)
```

```python
#Using Simple Polynomial Regression:
#keep on increasing degree and see if it fits
#applying polynomial transformation:
from sklearn.preprocessing import PolynomialFeatures
#polynomial features - is used to increase the degree/transformation.
poly = PolynomialFeatures(degree=2, include_bias=True)
#include_bias add x0 which is 1 (beta0)*1.
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test) #no reference to test data is taken.
```

## PIPELINE IN POLYNOMIAL:

```python
#create a generic function and depending on degree will see what kind of curve it fit in for new data.
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree): #degree is passed by frontend
  X_new = np.linspace(-6, 6, 200).reshape(200,1) #dataset
  poly_feat = PolynomialFeatures(degree=degree, include_bias=True) #new features in polynomail way
  lin_reg = LinearRegression() #model
  #both features and model should be added in pipeline one after another
  #1. feature
  #2. model and fit with data
  #combining this inside the pipeline.(in key value(object) pairs)
  poly_regression = Pipeline([
      ("poly_features", poly_feat),
      ("lin_reg", lin_reg)
  ])
  poly_regression.fit(X_train, y_train) #will transform based on degree
  y_pred_new = poly_regression.predict(X_new)
  #plotting prediction line:
  plt.plot(X_new, y_pred_new, "r", label="Degree "+str(degree), linewidth=4)
  plt.plot(X_train, y_train, "b.", linewidth=3, label="Training Data")
  plt.plot(X_test, y_test, "g.", linewidth=5, label="Testing Data")
  plt.legend(loc="upper right")
  plt.xlabel("X")
  plt.ylabel("y")
  plt.axis([-3,3,0,45])

```
