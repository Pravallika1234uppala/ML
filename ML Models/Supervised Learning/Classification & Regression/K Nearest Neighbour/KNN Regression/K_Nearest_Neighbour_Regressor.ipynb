## K NEAREST NEIGHBOUR REGRESSOR:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import make_regression

X, y = make_regression(
    n_samples=3000, #3000 observations or data points
    n_features=3, #3 input features
    noise=10, #Random Gaussian noise is added to y, making the relationship less
    #perfect (more like real-world data).
    random_state=42
)
```

```python
X
```

```python
y
```

```python
df1 = pd.DataFrame(X)
df2 = pd.DataFrame(y)
```

```python
df = pd.concat([df1, df2], axis=1, ignore_index=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.neighbors import KNeighborsRegressor

'''
n_neighbors(k)=5 default

weights-uniform and distance.
if all points weighted equally - uniform
if the weighting the points is inverse of their distance(so closer neighbours have greate influyence)


algorithm- need to find distance from new point to every other points to find 5 nearest neighbors.
These algorithms creates a binary tree.
so, due to decrease in search size bcz of trees Time complexity decreases and optimization increases.
can use:

ball_tree
kd_tree
brute
auto(based on the dataset any parameters are selected.)


p-2 (default and int)
This indicates. euclidean distance.
2-euclidean distance by default
if 1 - manhattan distance.

'''
```

```python
regmodel = KNeighborsRegressor(n_neighbors=7) #all default p=2, k=5, algorithm=auto
#if feastures are randomly distribution in features dimensions euclidean distance is prefereed.
```

```python
regmodel
```

```python
regmodel.fit(X_train, y_train)
```

```python
regmodel.get_params()
```

```python
y_pred = regmodel.predict(X_test)
```

```python
error = np.average(y_test-y_pred)
print(error)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print(f"Score: {r2_score(y_test, y_pred)}")
print(f"mean absolute error: \n{mean_absolute_error(y_test, y_pred)}")
print(f"mean square error: \n{mean_squared_error(y_test, y_pred)}")
```

### GRID SEARCH CV:

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
params = {
    'n_neighbors': [2,4,5,7,8,9],  # K values
    'weights': ['uniform', 'distance'],  # strategies of weights
    'metric': ['euclidean', 'manhattan', 'minkowski'],  # Distance measures
    'p': [1, 2]  # power parameter ( distance behavior)
}

grid_knr = GridSearchCV(KNeighborsRegressor(), param_grid=params, cv=5, scoring='accuracy', n_jobs=-1)
grid_knr.fit(X_train, y_train)


print(f"Best Parameters: {grid_knr.best_params_}")
print(f"Best Cross-Validation Accuracy: {grid_knr.best_score_:.4f}")


best_knn = grid_knr.best_estimator_
test_accuracy = best_knn.score(X_test, y_test)
print(f"Test Accuracy: {test_accuracy:.4f}")
```
