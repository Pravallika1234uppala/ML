```python
#dividing into independent and dependent features before encoding
#this is prevent test data leakage to training data.

X = df.drop('total_bill', axis=1)
y = df['total_bill']
```

```python
X
```

```python
y
```

```python
#now splitting training and testing data:
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train
```

```python
#feature encoding:
#for 2 categories - label encoding
#for more categories - one hot encoding

from sklearn.preprocessing import LabelEncoder, OneHotEncoder

#creating encoder based on features
lben1 = LabelEncoder()
lben2 = LabelEncoder()
lben3 = LabelEncoder()

```

```python
#to remove warnings:
import warnings
warnings.filterwarnings('ignore')
```

```python
X_train['sex']=lben1.fit_transform(X_train['sex'])
X_train['smoker']=lben2.fit_transform(X_train['smoker'])
X_train['time']=lben3.fit_transform(X_train['time'])

```

```python
X_train
```

```python
X_test['sex']=lben1.transform(X_test['sex'])
X_test['smoker']=lben2.transform(X_test['smoker'])
X_test['time']=lben3.transform(X_test['time'])
```

```python
X_test.head()
```

```python
#One hot Encoding(need to also use column transformer)
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

#to drop this encoding at first feature. and index where the column to encode is
#remainder to drop (to drop all remaining columns) other than encoded
#remainder to passthrough to leave all the other features.

#giving it as list [3] to transform easily
ct=ColumnTransformer(transformers=[('onehot', OneHotEncoder(drop='first'), [3])],
                                   remainder='passthrough')
```

```python
#to apply this colunm transformer to the data
import sys
import numpy as np

#to see the data clearly
np.set_printoptions(threshold=sys.maxsize)

ct.fit_transform(X_train)
```

```python
X_train=ct.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=ct.transform(X_test)
```

```python
X_test
```

```python
#SVR-Support Vector Regression:

from sklearn.svm import SVR

svr = SVR()
```

```python
svr
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
#error & performance metrics:
from sklearn.metrics import r2_score, mean_absolute_error

print("R2 score:",r2_score(y_pred,y_test))
print("Error:",mean_absolute_error(y_pred,y_test))
```

```python
#hyper parameter tuning

#using grid search cv

from sklearn.model_selection import GridSearchCV

#parameters range:
params = {'C': [0.1, 0.01, 1, 1.10, 10, 100, 1000],
          'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
          'kernel': ['rbf']
}
```

```python
grid_model = GridSearchCV(svr, param_grid=params, refit=True, verbose=3)

#fitting the grid model with svr as estimator
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_params_
```

```python
prediction_grid = grid_model.predict(X_test)
```

```python
#error & performance metrics:
from sklearn.metrics import r2_score, mean_absolute_error

print("R2 score:",r2_score(prediction_grid,y_test))
print("Error:",mean_absolute_error(prediction_grid,y_test))
```
