# LOGISTIC REGRESSION:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification #create a dataset with classification approiach
#doneed standardizattion as well- already created in such way
```

## BINARY LOGISTIC REGRESSION:

```python
#create a dataset:
X,y=make_classification(n_samples=1000,n_features=10,n_redundant=3,n_informative=2,n_classes=2,random_state=42)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
binlogmodel = LogisticRegression()

#can assign ridge and lasso for overfitting and feature selectiuon using penality parameter.
#C- inversly proportional to lambda.
#class weight - useful for imbalancve dataset.
```

```python
binlogmodel.fit(X_train,y_train)
```

```python
y_predbinlog = binlogmodel.predict(X_test)
```

```python
y_predbinlog
```

```python
binlogmodel.predict_proba(X_test)
#will give probability of a point to be in diff categoprys
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

X_df = pd.DataFrame(X_train)
for column in X_df.columns:
    plt.scatter(X_df[column], y_train)
    plt.scatter(X_df[column], binlogmodel.predict(X_train), color='red')
    plt.scatter(X_df[column], sigmoid(X_df[column]), color='k')
    plt.title(f'Logistic Regression plot for {column}')
    plt.show()
```

```python
Xt_df = pd.DataFrame(X_test)
for column in Xt_df.columns:
  plt.scatter(Xt_df[column], y_test)
  plt.scatter(Xt_df[column], binlogmodel.predict(X_test), color='red')
  plt.scatter(Xt_df[column], sigmoid(Xt_df[column]), color='k')
  plt.title(f'Logistic Regression plot for {column}')
  plt.show()
```

```python
#performance metrics:
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_test, y_predbinlog)
print(score)
```

```python
cm = confusion_matrix(y_test, y_predbinlog)
print(cm)
```

```python
print(classification_report(y_test, y_predbinlog))
```

## HYPER PARAMETER TUNING & CROSS VALIDATION:

```python
#1. Grid Search CV
#2.  Randomised Search CV
```

```python
#selecting diff parameters is possible in hyperparameter tuning.
model = LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
C_values = [100,10,1.0,0.1, 0.05, 1.98]#inverse to lambda
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
#these are main key parameter
#to pass this will have to make it key value pair
```

```python
params = dict(penalty=penalty, C=C_values, solver=solver)
```

```python
#now need to play with these to check for best parameters:
params
```

### 1. GRID SEARCH CV:

```python
#with this help will find best parameter that fits for dataset.
from sklearn.model_selection import GridSearchCV, StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=cv, n_jobs=-1)

#estimator - model used, param_grid - parameters used, scoring-accuracy in binary classification.
#cv used is Stratified k Fold
# n_jobs = -1 will lets you use all processors in the system.
```

```python
grid
```

```python
#This finds best combination and best parameters that fits the model.
grid.fit(X_train, y_train)
#finds best parameters for each parameter combination.
```

```python
grid.best_params_
```

```python
#these are best parameters that are selected.
grid.best_score_
```

```python
y_predgrid=grid.predict(X_test)
```

```python
score = accuracy_score(y_test, y_predgrid)
print(score)
cm = confusion_matrix(y_test, y_predgrid)
print(cm)
print(classification_report(y_test, y_predgrid))
```

```python
Xt_df = pd.DataFrame(X_test)
for column in Xt_df.columns:
  plt.scatter(Xt_df[column], y_test)
  plt.scatter(Xt_df[column], grid.predict(X_test), color='red')
  plt.title(f'Grid Search CV for {column}')
  plt.show()
```

```python
#here we did hyperparameter tuning with GridSearch Cv and cross validation called Stratified k fold.
```

### 2. Randomized Search CV

```python
#in grid search cv - it has combination of paraameters so more time
#in random search cv will pick random paramters
```

```python
from sklearn.model_selection import RandomizedSearchCV
model = LogisticRegression()
randomcv = RandomizedSearchCV(estimator=model, param_distributions=params, cv=5, scoring='accuracy')

```

```python
randomcv.fit(X_train, y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_estimator_
```

```python
randomcv.best_params_
```

```python
y_predrandcv = randomcv.predict(X_test)
```

```python
score = accuracy_score(y_test, y_predrandcv)
print(score)
cm = confusion_matrix(y_test, y_predrandcv)
print(cm)
print(classification_report(y_test, y_predrandcv))
```

```python
#for large dataset randomised cv is faster.
```
