##MULTI CLASS CLASSIFICATION PROBLEM

```python
'''
In Bernouli two categories of data can also be taken as 1 or 0 leading to a sparse matrix.

Sparse matrix - Entire Data has maximum of 1 and 0's.
This technique of bernouli is specifically used in NLP Problem statement.

and multinomial is specifically used in NLP Problem statements.

When we have Sparse matrix we can either use bernouli or Multinomial Naive Bayes.

So, for NLP can use multinomial and Bernouli Naive Bayes.

'''
```

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, y = load_iris(return_X_y=True)
# if return_X_y=True it says to return independent and dependent features.
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=4)
```

```python
X_train
```

```python
X_test
```

### GAUSSIAN NAIVE BAYES

```python
'''here input features are continuous. So, we use Gaussian Distribution.
For Categorical (binary) bernouli is preferred
For Text Input features Multinomial is preferred
'''
#here using Gaussian Naive Bayes
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb
```

```python
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
y_pred
```

```python
import numpy as np
error = np.average(y_test-y_pred)
print(error)
```

```python
y_pred_prob = gnb.predict_proba(X_test)
```

```python
y_pred_prob
```

```python
#To see plotting of train data:
import pandas as pd
import matplotlib.pyplot as plt
X_df = pd.DataFrame(X_train)
for column in X_df.columns:
  plt.scatter(X_df[column], y_train)
  plt.xlabel(f"Feature {column}")
  plt.ylabel("Output")
  plt.show()
```

```python
#performance metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

score = accuracy_score(y_test, y_pred)
print(score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
classreport = classification_report(y_test, y_pred)
print(classreport)
```

```python
plt.scatter(X_df.iloc[:,0], X_df.iloc[:,1], color="green")
plt.xlabel("Sepal width")
plt.ylabel("Sepal length")
plt.show()
```

```python
#plot for testing data:
Xt_df = pd.DataFrame(X_test)
for column in Xt_df.columns:
  plt.scatter(Xt_df[column], y_test, marker="*", color='blue')
  plt.scatter(Xt_df[column], gnb.predict(X_test), marker=".", color='red')
  plt.xlabel(f"Feature {column}")
  plt.ylabel("Output")
  plt.show()
```
