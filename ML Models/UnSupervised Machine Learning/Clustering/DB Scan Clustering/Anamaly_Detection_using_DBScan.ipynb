```python
from sklearn.cluster import DBSCAN
#from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
%matplotlib inline
```

```python
#make circles is to create a circles like data
#noise to add kind of deviations.

X,y = make_circles(n_samples=800, factor=0.3, noise=0.1)
```

```python
X.shape
```

```python
y.shape
```

```python
#plotting data
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
dbscan_model=DBSCAN(eps=0.1)
#parameters:radius
```

```python
dbscan_model
```

```python
dbscan_model.fit_predict(X)
```

```python
dbscan_model.labels_
#this is y label
```

```python
#plotting with predicted
plt.scatter(X[:,0],X[:,1],c=dbscan_model.labels_)
```
