```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
#importing iris dataset:

iris_data = datasets.load_iris()
```

```python
iris_data
```

```python
iris_data.data
```

```python
iris_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
```

```python
iris_df.head()
```

```python
#scaling:
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(iris_df)
```

```python
X_scaled
```

```python
#decreasing dimensions - PCA

from sklearn.decomposition import PCA

pca=PCA(n_components=2)

pca_scaled = pca.fit_transform(X_scaled)
pca_scaled
#converted to 2 d
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=iris_data.target)

```

```python
#agglomerative clustering - by constructing dendograms

import scipy.cluster.hierarchy as sc

#plotting
plt.figure(figsize=(20,8))
plt.title("Dendograms")

#creating dendograms: (using linkage model for scipy and different methods)

sc.dendrogram(sc.linkage(pca_scaled, method='ward'))


plt.title("Dendogram:")
plt.xlabel('Sample Index')
plt.ylabel('Eucledian Distance')
```

```python
from sklearn.cluster import AgglomerativeClustering

#doing clustering for above
cluster = AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')

cluster.fit(pca_scaled)

```
