```python
from sklearn.cluster import DBSCAN
#from sklearn.datasets import make_moons
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
%matplotlib inline
```

```python
#make circles is to create a circles like data
#noise to add kind of deviations.

X,y = make_circles(n_samples=800, factor=0.4, noise=0.2)
```

```python
X
```

```python
X.shape
```

```python
y.shape
```

```python
df=pd.DataFrame(X)
```

```python
df
```

```python
#plotting data
plt.scatter(df.iloc[:,0], df.iloc[:,1], c=y)
```

```python
from sklearn.ensemble import IsolationForest
```

```python
If = IsolationForest(contamination=0.2)
If.fit(df)
predictions = If.predict(df)
```

```python
predictions
```

```python
ana_index = np.where(predictions<0)
ana_index
```

```python
x=df.values
```

```python
x
```

```python
#plotting with predicted
plt.scatter(df.iloc[:,0],df.iloc[:,1])
plt.scatter(x[ana_index,0], x[ana_index,1], edgecolors="r")
```
